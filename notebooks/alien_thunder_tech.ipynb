```python
import pandas as pd
from pathlib import Path
X_train = pd.read_csv(Path(r"C:\Users\90553\Documents\ml_data_analysis\data\X_train.csv"),sep=r"\s+")
Y_train = pd.read_csv(Path(r"C:\Users\90553\Documents\ml_data_analysis\data\Y_train.csv"),header=None, names=['category'])
X_test = pd.read_csv(Path(r"C:\Users\90553\Documents\ml_data_analysis\data\X_test.csv"),sep=r"\s+")
Y_test = pd.read_csv(Path(r"C:\Users\90553\Documents\ml_data_analysis\data\Y_test.csv"),header=None, names=['category'])
print(X_train.shape, Y_train.shape, X_test.shape, Y_test.shape)
```

```python
XY_train = pd.concat([X_train, Y_train], axis=1)
print(XY_train.shape)
XY_train.head()
```

```python
word = "alien"
counts = XY_train.groupby('category')[word].sum().reset_index()
counts.columns = ['category', 'count']
print(counts)
```

```python
word = "thunder"
counts = XY_train.groupby('category')[word].sum().reset_index()
print(counts.head())
counts.columns = ['category', 'count']
```

```python
selected = XY_train.loc[:, ['category', 'thunder', 'alien']]
print(selected.head())
counts = selected.groupby('category').sum()
print(counts)
```

```python
# assuming last column is 'category'
word_cols = XY_train.columns.drop('category')

counts = XY_train.groupby('category')[word_cols].sum().sum(axis=1).reset_index(name='total_word_count')
print(counts)
```
